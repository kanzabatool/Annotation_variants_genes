```python
import pandas as pd

def load_data():
   return pd.read_csv("prostate.csv")

data = load_data()
```

```python
data
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df=pd.DataFrame(data)
```

```python
X=df[['Unnamed: 0','lcavol','lweight','age','lbph','svi','lcp','gleason','pgg45']]
y=df['lpsa']
```

Linear Regression 

```python
#baseline model

regression=LinearRegression()
regression.fit(X,y)
first_model=(mean_squared_error(y_true=y,y_pred=regression.predict(X)))
print(first_model)
```

Ridge Regression

```python
ridge=Ridge(normalize=True)
search=GridSearchCV(estimator=ridge,param_grid={'alpha':np.logspace(-5,2,8)},scoring='neg_mean_squared_error',n_jobs=1,refit=True,cv=10)
```

```python
search.fit(X,y)
search.best_params_
#{'alpha': 0.01}
```

```python
abs(search.best_score_)
```

```python
ridge=Ridge(normalize=True,alpha=0.01)
ridge.fit(X,y)
second_model=(mean_squared_error(y_true=y,y_pred=ridge.predict(X)))
print(second_model)
```

All subset Regression

```python
import numpy as np
import pandas as pd
import urllib
from itertools import chain, combinations
import statsmodels.api as sm
```

```python

n_features = X.shape[1]
subsets = chain.from_iterable(combinations(range(n_features), k+1) for k in range(n_features))
best_score = -np.inf
best_subset = None
for subset in subsets:
    #print(subset)
    lin_reg = sm.OLS(y, X.ix[:, subset]).fit()
    score = lin_reg.rsquared_adj  #Adjusted R square
    if score > best_score:
        best_score, best_subset = score, subset
print(best_subset, best_score)
```
